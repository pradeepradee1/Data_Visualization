# 3.Subplots

```python
plt.subplot(2,2,1)
# Here Subplot describe is Two row and two Columns and places it in 1 position
plt.plot(x,y,'r')
plt.subplot(2,2,2)
# Here Subplot describe is Two row and two Columns and places it in 2 position
plt.plot(x,y,'b')
plt.subplot(2,2,3)
# Here Subplot describe is Two row and two Columns and places it in 3 position
plt.plot(x,y,'g')
plt.subplot(2,2,4)
# Here Subplot describe is Two row and two Columns and places it in 4 position
plt.plot(x,y,'y')
```

```python
x= np.arange(0, 4 * np.pi, 0.1)
y= np.sin(x)
x,y
```

```python
plt.title('sin wave form')
plt.plot(x,y)
```

```python
x= np.arange(0, 4 * np.pi, 0.1)
y_sin= np.sin(x)
y_cos= np.cos(x)
x,y_sin,y_cos
```

```python
plt.subplot(2,1,1)
# Here Subplot describe is Two row and Single Columns
plt.plot(x,y_sin)
plt.title("sine")

plt.subplot(2,1,2)
# Here Subplot describe is Two row and Single Columns
plt.plot(x,y_cos)
plt.title("cosine")
```

# 5.Histgrams 

```python
# Here is Histograms 
# 1) X axis describes the array of a 
# 2) Y axis describes the density or Count of particular array of a 
```

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("Histogram")
```

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
print(a)
plt.hist(a,bins=20)
plt.title("Histogram")
```

# 6.Box plot 

```python
data= [ np.random.normal(0,std,100) for std in range(1,4)]
data
```

```python
plt.boxplot(data,vert=True,patch_artist=True)
```

```python
plt.boxplot(data,vert=False,patch_artist=True)
```

# 7.PIE chart

```python
labels = 'python',"C++","Ruby","java"
sizes = [500,130,245,210]
colors= ["gold","yellowgreen","lightcoral","lightskyblue"]
explode = (0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

```python
labels = 'python',"C++","Ruby","java"
sizes = [500,130,245,210]
colors= ["gold","yellowgreen","lightcoral","lightskyblue"]
explode = (0.5,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

```python
labels = 'python',"C++","Ruby","java"
sizes = [500,130,245,210]
colors= ["gold","yellowgreen","lightcoral","lightskyblue"]
explode = (0.1,2,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```
