```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ["Google","Amz","Msft","Fb"]
revenue = [90,136,89,27]
```

```python
plt.bar(company,revenue)
```

```python
plt.ylabel("revenue")
plt.title("US Tech Stocks")
plt.bar(company,revenue ,label="Revenue")
plt.legend()
```
