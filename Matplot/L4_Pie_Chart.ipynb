```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels = ["Home_Rent","Food","Phone/Internet bill","car","other utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.2f%%")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.f%%")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.2f%%",shadow=True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.2f%%",explode=[0,0.2,0,0,0])
plt.show()
```
