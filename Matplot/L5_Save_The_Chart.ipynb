#### 1 ) using plt.savefig

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels = ["Home_Rent","Food","Phone/Internet bill","car","other utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
plt.savefig("piechart.png")
```

```python
plt.pie(exp_vals,labels=exp_labels)
plt.savefig("piechart1.png",bbox_inches="tight",pad_inches=2,transparent=True)
```

#### 2) using pickle (Save and Open)

```python
import pickle
```

```python
plot=plt.pie(exp_vals,labels=exp_labels,autopct="%0.2f%%")
pickle.dump(plot, open("plot1.pickle", "wb"))
```

```python
plt = pickle.load(open("plot1.pickle", "rb"))
```
